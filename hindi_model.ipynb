```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("IMDB-Movie-Dataset(2023-1951).csv")
movies.head(1)
```

```python
#lets summorise our selected data
#genres
#id
#keywords
#title
#overview
#cast
#crew
pross_movie = movies[['movie_id','movie_name','overview', 'genre','director',]]
```

```python
#checking for null data
pross_movie.isnull().sum()
```

```python
pross_movie['overview'] = pross_movie['overview'].apply(lambda x: x.split())
pross_movie['director'] = pross_movie['director'].apply(lambda x: x.split())
pross_movie['genre'] = pross_movie['genre'].apply(lambda x: x.split())
```

```python
pross_movie['genre'] = pross_movie["genre"].apply(lambda x:[i.replace(" ","") for i in x])
pross_movie['director'] = pross_movie["director"].apply(lambda x:[i.replace(" ","") for i in x])
pross_movie['overview'] = pross_movie["overview"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
pross_movie['tags'] = pross_movie['overview'] + pross_movie["genre"]+ pross_movie["director"]
```

```python
data_set = pross_movie[['movie_id',"movie_name","tags"]]
```

```python
data_set.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return  " ".join(y)
```

```python
data_set['tags'] = data_set['tags'].apply(lambda x:" ".join(x))
data_set['tags'][0]
```

```python
data_set['tags'] = data_set['tags'].apply(lambda x:x.lower())
```

```python
data_set['tags']=data_set['tags'].apply(stem)
```

```python
data_set['tags'][0]
```

```python
data_set['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
vectors = cv.fit_transform(data_set['tags']).toarray()
# converted the data set in the form of vectors
vectors
```

```python
features_name = cv.get_feature_names_out()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[2]
```

```python
def recommend(movie):
    movie_index = data_set[data_set['movie_name'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(data_set.iloc[i[0]].movie_name)
```

```python
recommend("Jawan")
```
